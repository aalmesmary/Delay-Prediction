```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from datetime import datetime
```

```python
# Load the datasets
dataset1 = pd.read_excel("../data/raw/ESTIDAMA PKG-3 Schedule Updates Oct-24.xlsx")  # Replace with the actual path
dataset2 = pd.read_excel("../data/raw/P2543-R01-Oct24. Rel.xlsx", sheet_name="TASK")  # Replace with the actual path
dataset3 = pd.read_excel("../data/raw/P2543-R01-Oct24. Rel.xlsx", sheet_name="TASKPRED")  # Replace with the actual path
```

```python
#remove spaces
dataset1['Activity ID']=dataset1['Activity ID'].str.strip()
dataset2['task_code']=dataset2['task_code'].str.strip()
dataset3['task_id']=dataset3['task_id'].str.strip()
```

```python
merged_df = pd.merge(dataset1, dataset2, how='left', left_on='Activity ID', right_on='task_code')
merged_df = pd.merge(merged_df, dataset3, how='left', left_on='Activity ID', right_on='task_id')
```

```python
merged_df
```

```python
completed_df = merged_df[merged_df['status_code'] == 'Completed']
```

```python
completed_df.info()
```

```python
completed_df['Activity ID'].duplicated().sum()
```

```python
completed_df.dropna(subset=['Baseline Start','Baseline Finish', 'Start', 'Finish', 'start_date', 'end_date'], inplace=True)
```

```python
completed_df.info()
```

```python
completed_df[['Baseline Start','Baseline Finish', 'Start', 'Finish', 'start_date', 'end_date']].sample(10)
```

```python
completed_df['end_date'] = pd.to_datetime(completed_df['end_date'], format='%Y-%m-%d %H:%M:%S')
```

```python
# Calculate delay duration
completed_df['Delay_Duration'] = (completed_df['end_date'] - completed_df['Baseline Finish']).dt.days
```

```python
completed_df['Delay_Duration'].describe()
```

```python
# Define conditions
conditions = [
    completed_df['Delay_Duration'] == 0,
    completed_df['Delay_Duration'] < 0,
    completed_df['Delay_Duration'] > 0
]

# Define corresponding choices
choices = ['On time', 'Earlied', 'Delayed']

# Use numpy.select to create 'Delay_Status'
completed_df['Delay_Status'] = np.select(conditions, choices)
```

```python
completed_df[['Delay_Status']].value_counts()
```

```python
# Feature engineering: Count predecessors/successors
completed_df['Predecessor_Count'] = completed_df['pred_list'].apply(lambda x: len(x.split(',')) if pd.notna(x) else 0)
completed_df['Successor_Count'] = completed_df['succ_list'].apply(lambda x: len(x.split(',')) if pd.notna(x) else 0)
```

```python
# Encode categorical variables
le = LabelEncoder()
completed_df['Delay_Status'] = le.fit_transform(completed_df['Delay_Status'])
```

```python
completed_df['Delay_Status'].value_counts()
```

```python
# Save the data
completed_df.to_csv('../data/completed_tasks.csv', index=False)
```
